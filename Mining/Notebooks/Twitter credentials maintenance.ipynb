```python
def make_query(terms):
    """
    >>>make_query(['pain', 'fibromyalgia', 'migraine'])
        pain%2C OR fibromyalgia%2C OR migraine

    :return:
    """
    seperator = '%2C OR '
    stem = "q="
    last = terms[-1:][0]
    
    for term in terms:
        stem += "%s" % term
        if term != last:
            stem+= seperator
    return stem



terms = ['pain', 'fibromyalgia', 'migraine']
assert (make_query(terms) ==  "q=pain%2C OR fibromyalgia%2C OR migraine")
# make_query(terms)
```

```python


def convert_object_into_dict( result ):
    """The search script returns objects, converts into a dictionary so
    that they can be serialized when we send them to ther server.
      """
    def process(val):
        if isinstance(val, object):
            # recursively call
            val = convert_object_into_dict(val)
#         if type(val) is list:
            # even more recursively call
#             val = [ process(v) for v in val ]
        return val

    if type( result ) is dict: return result  # just in case...
    return { k: process(getattr( result, k  )) for k in result.__dict__ }
```
